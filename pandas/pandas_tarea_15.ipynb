```python
df_N= pd.DataFrame({'Reales': 1.,
                    'Series': pd.Series(1, index=list(range(4)), dtype='float32'),
                    'Arrays 1D': np.array([9999] * 4, dtype='int32'),
                    'Fechas': pd.Timestamp('20200110'),
                    'Categoricas': pd.Categorical(["Validacion", "test", "test", "train"]),
                    'Strings':'España'})

df_N
```

```python
print(type(df_N))
print(type(df_N["Reales"]))
print(type(df_N["Reales"][0]))
print(type(df_N["Arrays 1D"][0]))
print(type(df_N["Fechas"][0]))
print(type(df_N["Categoricas"][0]))
```

```python
df_climate1 = pd.read_csv("./data_files/GlobalLandTemperatures_GlobalLandTemperaturesByMajorCity_jgx.csv")
df_climate1
```

```python
#Know the size of the database
df_climate1.size
```

```python
#Get the name of the variables
df_climate1.columns
```

```python
#Delete the first column
df_climate1.drop(df_climate1.columns[df_climate1.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df_climate1
```

```python
#Eliminate values that are not a number.
df_climate1.dropna(how="any", inplace = True)
df_climate1
```

```python
#Check if there is any difference with the lenght of the previous df
df_climate1.size
```

```python
#Show the last 6 rows of the df.
df_climate1.tail(6)
```

```python
df_climate3 = df_climate1.copy()
df_climate3
```

```python
# Crea un nuevo data frame denominado ``df_climate3_country`` a partir de ``df_climate3`` 
# que contenga las muestras climáticas correspondientes a la ciudad de Xian.

#city_col = df_climate3.loc[:,"City"]
value_xian = df_climate3.loc[df_climate3.City == "Xian"]


df_climate3_country = value_xian
df_climate3_country
```

```python
# Convierte la variable ``dt`` en una variable de tipo ``datetime`` con formato "%Y-%m-%d".

df_climate3.dt = pd.to_datetime(df_climate3.dt, format="%Y-%m-%d")
df_climate3.dtypes
```

```python
# Obtén los estadísticos descriptivos de las variables numéricas del dataframe 
# ``df_climate3`` (media, min, max, cuartiles 1 2 y 3, etc..).
media_av = df_climate3.AverageTemperature.median()
print(f"Median Average Temperatures: {media_av}")
media_avu = df_climate3.AverageTemperatureUncertainty.median()
print(f"Median Average Temperatures Uncertainty: {media_avu}")

min_av = df_climate3.AverageTemperature.min()
print(f"Min Average Temperatures: {min_av}")
min_avu = df_climate3.AverageTemperatureUncertainty.min()
print(f"Min Average Temperatures Uncertainty: {min_avu}")

max_av = df_climate3.AverageTemperature.max()
print(f"Max Average Temperatures: {max_av}")
max_avu = df_climate3.AverageTemperatureUncertainty.max()
print(f"Max Average Temperatures Uncertainty: {max_avu}")

quantile_av1 = df_climate3.AverageTemperature.quantile(q=0.1)
print(f"1% Quantile Average Temperatures: {quantile_av1}")
quantile_avu1 = df_climate3.AverageTemperatureUncertainty.quantile(q=0.1)
print(f"1% Quantile Average Temperatures Uncertainty: {quantile_avu1}")

quantile_av2 = df_climate3.AverageTemperature.quantile(q=0.2)
print(f"2% Quantile Average Temperatures: {quantile_av2}")
quantile_avu2 = df_climate3.AverageTemperatureUncertainty.quantile(q=0.2)
print(f"2% Quantile Average Temperatures Uncertainty: {quantile_avu2}")

quantile_av3 = df_climate3.AverageTemperature.quantile(q=0.3)
print(f"3% Quantile Average Temperatures: {quantile_av3}")
quantile_avu3 = df_climate3.AverageTemperatureUncertainty.quantile(q=0.3)
print(f"3% Quantile Average Temperatures Uncertainty: {quantile_avu3}")
```

```python
# Crea un nuevo dataframe denominado ``df_climate4`` a partir de una copia de ``df_climate3`` y cuyo índice 
# (nombre de la muestra) esté constituído por los valores de la variable ``dt``. 
df_climate4=df_climate3.copy()
df_climate4 = df_climate4.set_index("dt")
df_climate4
```

```python
# Elimina las muestras del data frame ``df_climate4`` anteriores a la fecha "1900-1-2".
val_to_delete = df_climate4.loc[df_climate4.index >= "1900-1-2"]
val_to_delete
```
