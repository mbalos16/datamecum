```python
#6.1 6.2, 6.3, 6.4
data = pd.read_csv("./data_files/international-airline-passengers.csv")
time_df = DataFrame(data)
time_df[:8]

#Tansform the series Month into datetime format.
time_df.Month = pd.to_datetime(time_df.Month, format = "%Y-%m")

#Set the Month column as the index of the sataframe.
time_df = time_df.set_axis(time_df.Month)
del time_df["Month"]
time_df
```


* Representa la serie temporal utilizando matplotlib. Te atreves a hacer lo mismo utilizando el método ``plot`` del dataframe?

```python
fig, ax = plt.subplots()
fig.set_size_inches(18,6)
ax.plot(time_df.index, time_df["Passengers"],  marker="o", color = "red")

#Add labels
ax.set_xlabel("Years")
ax.set_ylabel("Passangers")
ax.set_title("International airline passangers over the years")

plt.plot()
plt.show()
```

* Realiza un análisis sencillo de series temporales utilizando la función ``tsa.seasonal_decompose`` que hemos visto en el ejemplo anterior. ¿Observa algún tipo de estacionalidad en las serie? 

```python
import statsmodels.api as sm
decompose_time = sm.tsa.seasonal_decompose(time_df["Passengers"], model="multiplicative")

fig = decompose_time.plot()
fig.set_size_inches(10,7)
```

```python
aprox = decompose_time.trend * decompose_time.seasonal
fig, ax = plt.subplots()
fig.set_size_inches(10,7/4)
ax.plot(aprox, label = "prediction")
ax.plot(time_df.index, time_df["Passengers"], label = "target")
ax.legend()
plt.show()
```

```python
error = time_df["Passengers"] - aprox
fig, ax = plt.subplots()
fig.set_size_inches(10,7/4)
ax.plot(error, label = "error")
ax.legend()
plt.show()
```
